```python
import pandas as pd
import numpy as np
```

```python
datacrime=pd.read_csv("datacrime.csv")
datacrime
```

```python
#mengetahui banyak data missing value
np.sum(datacrime.isnull())
```

```python
#statistika deskriptif tiap variabel
datacrime.describe()
```

```python
#mengatasi missing value menggunakan nilai mean
datacrime1=datacrime
datacrime1['predik1']=datacrime['predik1'].fillna((datacrime1['predik1'].mean()))
datacrime1['predik2']=datacrime['predik2'].fillna((datacrime1['predik2'].mean()))
datacrime1['predik3']=datacrime['predik3'].fillna((datacrime1['predik3'].mean()))
datacrime1
```

```python
#menyimpan file data tanpa missing value
from pandas import DataFrame
datacrimetmv = pd.DataFrame(datacrime1,columns=['respon','predik1','predik2','predik3'])
datacrimetmv.to_excel ("datacrime_tanpa_mv.xlsx")
```

```python
#mengatasi outlier
datacrime2=datacrime1
from scipy import stats
z = np.abs(stats.zscore(datacrime2._get_numeric_data()))
print(z)
datacrime2 = datacrime2[(z < 3).all(axis=1)]
print(datacrime.shape)
datacrime2
```

```python
#menyimpan file data tanpa missing value dan outlier
from pandas import DataFrame
datacrimenoout = pd.DataFrame(datacrime2,columns=['respon','predik1','predik2','predik3'])
datacrimenoout.to_excel ("datacrime_tanpa_outlier.xlsx")
```

```python
#normalisasi data yang belum diatasi outliernya
from sklearn.preprocessing import normalize
datacrime3=normalize(datacrime1,norm='l2') #default norm L2, other are L1 and max
datacrime3=pd.DataFrame(datacrime3)
datacrime3
```

```python
#transformasi datacrime3 menggunakan metode 3sigmoidal
datacrime3trans=(1-2.718281828**(datacrime3))/(1+2.718281828**(datacrime3))
datacrime3trans
```

```python
#menyimpan file data transformasi with outlier
from pandas import DataFrame
datacrimenormout = pd.DataFrame(datacrime3trans)
datacrimenormout.to_excel ("datacrime_trans_outlier.xlsx")
datacrimenormout
```

```python
#normalisasi data yang sudah diatasi outliernya
from sklearn.preprocessing import normalize
datacrime4=normalize(datacrime2,norm='l2') #default norm L2, other are L1 and max
datacrime4=pd.DataFrame(datacrime4)
datacrime4
```

```python
#transformasi datacrime4 menggunakan metode 3sigmoidal
datacrime4trans=(1-2.718281828**(datacrime4))/(1+2.718281828**(datacrime4))
datacrime4trans
```

```python
#menyimpan file data normalisasi no outlier
from pandas import DataFrame
datacrimenorm = pd.DataFrame(datacrime4trans)
datacrimenorm.to_excel ("datacrime_transformasi.xlsx")
```
