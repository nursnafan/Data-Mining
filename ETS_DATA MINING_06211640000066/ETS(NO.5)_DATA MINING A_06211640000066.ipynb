```python
#General Package
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
#membuka file data dengan format csv
retail = pd.read_csv('online_retail.csv')
#mengetahui 5 data teratas
retail.head()
```

```python
#mengetahui ukuran baris dan kolom dari data
retail.shape
```

```python
#mendeteksi jumlah missing pada tiap variabel
np.sum(retail.isnull())
```

```python
#melakukan cleaning data yang kosong dan menghilangkan transaksi credit serta tidak ada penjualan
retail['UnitPrice'] = retail['UnitPrice'].astype('str')
retail = retail[~retail['UnitPrice'].str.contains('0')]
```

```python
#mendeteksi jumlah missing pada tiap variabel
np.sum(retail.isnull())
```

```python
#mengelompokkan data berdasarkan variabel InvoiceNo
online_retail = retail.groupby('InvoiceNo')
```

```python
#merekam data dalam bentuk kolom
records = []
for i,j in online_retail:
    records.append(list(j['Description'].map(str)))
```

```python
#data setelah direkam dalam bentuk kolom
records
```

```python
#melakukan analisis association rules
association_rules=apriori(records,min_support=0.0045,
                          min_confidence=0.2,
                          min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
#menampilkan hasil analisis association rules
association_result
```

```python
len('association rules')
```

```python
#merapikan hasil dari association_rules
for item in association_result:
#first index of the inner list
#contains base item and add item 
    pair = item[0]
    items = [x for x in pair]
    print("rule: "+ items [0] + " -> " + items [1])
    
    #second index of the inner list
    print("support: " + str(item[1]))
    
    #third index of the list located at 10th
    #of the third index of the inner list
    print("confidence: " + str(item[2][0][2]))
    print("lift: "+ str(item[2][0][3]))
    print("=======================================")
```
